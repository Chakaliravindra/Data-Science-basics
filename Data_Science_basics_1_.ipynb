 1.1).sub-plot (adding sub-plot)
* -----1row,2columns-------

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(1,2,1)   # (1row,2columns,1column)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)   # (1row,2columns.2column)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

* -----2rows,1column-------

```python
plt.subplot(2,1,1)   # (2row,1columns,1st row)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(2,1,2)   # (2row,1columns.2nd row)
plt.plot(x,y2,color='r',linestyle=':',linewidth=2)

plt.show()
```

3).Scatter-plot

* changing mark Aesthetics

```python
x=[1,2,3,4,5,6,7,8,9]
y=[4,7,5,2,9,6,1,3,8]
plt.scatter(x,y,marker='*',c='red',s=100)  # (either red or r), "s" is the size of intersecting point
plt.show()
```

* adding two markers in the same plot

```python
x=[1,2,3,4,5,6,7,8,9]
y1=[4,7,5,2,9,6,1,3,8] 
y2=[9,4,5,3,2,8,6,1,4]

plt.scatter(x,y1,marker='*',c='g',s=100) 
plt.scatter(x,y2,marker='.',c='r',s=300)  
plt.show()
```

adding sub-plots
* adding two markers in the same plot

```python
x=[1,2,3,4,5,6,7,8,9]
y1=[4,7,5,2,9,6,1,3,8] 
y2=[9,4,5,3,2,8,6,1,4]

plt.subplot(1,2,1)
plt.scatter(x,y1,marker='*',c='g',s=100) 

plt.subplot(1,2,2)
plt.scatter(x,y2,s=300)  

plt.show()
```
